### Questão 1
Uma empresa fabricante de pneus automotivos descobriu que a vida útil dos pneus segue uma distribuição normal com média de 40.000 km e desvio padrão de 3.000 km. Acredita-se que uma mudança no processo de produção resultará em um produto melhor e, por isso, a empresa desenvolveu um novo pneu. Uma amostra de 100 pneus novos foi selecionada. A empresa constatou que a vida útil média desses novos pneus é de 40.900 km. Pode-se concluir que o novo pneu é significativamente melhor que o antigo, usando um nível de significância de 0,01?

H0: μ <= 40.000 km \
HA: μ > 40.000 km

```python
Zcal = (40900 - 40000)/(3000/np.sqrt(100))
Zcal
```

```python
Zcrit = norm.ppf(0.99)
Zcrit
```

Conclusão:
Aqui temos um teste Z de uma cauda (à direita). Como temos Zcal > Zcrit, rejeitamos a hipótese nula. Portanto, o novo pneu é significativamente melhor que o pneu antigo.

### Questão 2
Sabe-se que a alcalinidade, em miligramas por litro, da água nos trechos superiores dos rios de uma determinada região segue uma distribuição normal com um desvio padrão de 10 mg/l. As leituras de alcalinidade nos trechos inferiores dos rios da mesma região também seguem uma distribuição normal, mas com um desvio padrão de 25 mg/l. Dez leituras de alcalinidade são feitas nos trechos superiores de um rio da região e quinze nos trechos inferiores do mesmo rio, com os seguintes resultados.

trechos_sup = [91, 75, 91, 88, 94, 63, 86, 77, 71, 69] \
trechos_inf = [86, 95, 135, 121, 68, 64, 113, 108, 79, 62, 143, 108, 121, 85, 97]

Investigue, com um nível de confiança de 95%, a afirmação de que a verdadeira alcalinidade média da água no curso inferior deste rio é maior do que no curso superior.

H0: μinf <= sup \
HA: μinf > μsup 

```python
trechos_sup = [91, 75, 91, 88, 94, 63, 86, 77, 71, 69] 
trechos_inf = [86, 95, 135, 121, 68, 64, 113, 108, 79, 62, 143, 108, 121, 85, 97]

stats.ttest_ind(trechos_inf, trechos_sup, equal_var=False, alternative='greater')
```

Conclusão: \
Temos pValue < 0.05, assim, rejeitamos a hipótese nula. Portanto, a afirmação de que a verdadeira alcalinidade média da água no curso inferior do rio é maior do que no curso superior está correta.

Pedimos a 10 alunos estudem em silêncio no primeiro semestre e estudem ouvindo música clássica no segundo semestre, e registramos o número médio de horas gasto executando trabalhos por semana em cada semestre. Os tempos gastos na execução de trabalhos em 10 semanas de estudo estão listados a seguir. Podemos concluir que estudar ouvindo música clássica reduz o número de horas gastas por semana na execução dos trabalhos escolares? Analise o valor de t_calc (statistic) com confiança de 95%.

Importante: Sua conclusão não deve utilizar o pvalue. Conclua utilizando o valor de t_calc e o t_crítico obtido da tabela t. Note que não é necessário fazer o cálculo à mão, já que o t_calc (statistic) é fornecido pelo Python.

Em silêncio: 14, 13, 16, 21, 15, 19, 11, 20, 19, 16 \
Ouvindo música: 12, 13, 15, 22, 16, 19, 8, 17, 18, 17

H0: μsilencio <= μmusica \
Ha : μsilencio > μmusica

```python
p_silencio = [14, 13, 16, 21, 15, 19, 11, 20, 19, 16]
p_musica = [12, 13, 15, 22, 16, 19, 8, 17, 18, 17]
stats.ttest_rel(p_silencio, p_musica, alternative='greater')
```

Note que o valor de pValue reafirma essa conclusão.

### Questão 6
Uma turma do quarto ano está estudando o meio ambiente. Uma das tarefas é cultivar feijões em diferentes tipos de solo. Pedro escolheu cultivar seus feijões em terra encontrada do lado de fora da sala de aula, misturada com fiapos de secadora. Ana escolheu cultivar seus feijões em terra para vasos comprada em um viveiro local. Ivo escolheu cultivar seus feijões em terra do jardim de sua mãe. Nenhum produto químico foi usado nas plantas, apenas água. Elas foram cultivadas dentro da sala de aula, perto de uma grande janela. Cada criança cultivou cinco plantas. Ao final do período de crescimento, cada planta foi medida, registrando-se os dados (em polegadas).

pedro = [24, 21, 23, 30, 23] \
ana = [25, 31, 23, 20, 28] \
ivo = [23, 27, 22, 30, 20]

Faça um teste de hipótese capaz de avaliar a se o tipo de solo fez alguma diferença no crescimento do feijão. Utilize confiança de 95%.

H0: μpedro = μana = μivo \
Ha: Médias não são iguais

```python
pedro = [24, 21, 23, 30, 23] 
ana = [25, 31, 23, 20, 28] 
ivo = [23, 27, 22, 30, 20]

stats.f_oneway(pedro, ana, ivo)
```

Conclusão: \
Temos pValue > 0.05, portanto falhamos em rejeitar a hipótese nula. Não há evidências que o tipo de solo faz diferença no crescimento do feijão.
