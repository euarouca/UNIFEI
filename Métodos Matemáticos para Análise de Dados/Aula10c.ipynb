## Gráfico de Barras
Vamos construir um gráfico de barras usando os dados do dataset Ebola.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ebola.csv')
```

```python
df
```

```python
plt.rcParams['figure.figsize'] = [8,6]
sns.barplot(x='Type', y='Death', hue='Countries', data=df, errorbar=None)
plt.title('Mortalidade do Ebola nos países Serra Leoa e Guiné')
plt.xlabel('Tipos de casos')
plt.ylabel('Número de mortes')
plt.show()
```

Vamos agora trabalhar com o dataset Migrants.csv. O primeiro passo é entender esse dataset, então vamos exibir alguns registros inicialmente.

```python
mexicanos = pd.read_csv('migrants.csv', encoding='latin-1')
```

```python
mexicanos
```

Vamos ver as diferentes regiões onde os incidentes aconteceram.

```python
mexicanos['Region_of_Incident'].unique()
```

Queremos selecionar todas as regiões de incidente, com exeção da North America, South America, East Asia and South Asia.


```python
df2 = mexicanos[~mexicanos['Region_of_Incident'].isin(['North America', 'South America', 'South Asia', 'East Asia'])]
df2
```

Agora vamos criar um gráfico de barras mostrando o número de mortos e desaparecidos no ano de 2016.

```python
df3 = pd.crosstab(df2['Region_of_Incident'], df2['Reported_Year'], values=df2['Total_Dead_and_Missing'], aggfunc=np.sum)
df3
```

```python
plt.rcParams['figure.figsize'] = [10,8]
sns.set_context('notebook', font_scale=1.5, rc={'font.size': 14, 
                                                'axes.titlesize': 16,
                                                'axes.labelsize': 12})
df3.plot(kind='bar', stacked=False)
plt.title('Migrante mortos ou desaparecidos de 2014 a 2017')
plt.xlabel('Regiao')
plt.ylabel('Mortos ou desaparecidos')
plt.show()
```
