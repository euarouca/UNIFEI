### Questão 01
O índice de felicidade é uma métrica utilizada para mensurar o critério subjetivo "felicidade" nos diferentes países do globo. Utilizando o dataset 'happiness_2016', teste a hipótese de que as pessoas dos países da europa ocidental (Western Europe) são igualmente felizes às pessoas da América Latina e Caribe (Latin America and Caribbean). Nível de confiança: 95%.

H0: μ1 = μ2 \
Ha: μ1 ≠ μ2

```python
df = pd.read_csv('happiness_2016.csv')
df.head()
```

```python
df['Region'].unique()
```

```python
df1 = df[df['Region'] == 'Western Europe']
df2 = df[df['Region'] == 'Latin America and Caribbean']
```

```python
stats.ttest_ind(df1['Happiness Score'], df2['Happiness Score'], equal_var=True)
```

Conclusão: \
Temos que pValue < 0.05, rejeitamos a hipótese nula. Portanto, as pessoas da Europa Ocidental e da América Latina e Caribe Não são igualmente felizes.

### Questão 02
Dinheiro traz felicidade? Para responder a essa pergunta, realize um teste de hipótese com os 20 países mais ricos e os 20 países mais pobres do relatório de felicidade de 2016. Use a coluna Economy (GDP per Capita). Nível de confiança: 95%.

H0: μ_ricos <= μ_pobres \
Ha: μ_ricos > μ_ricos \
OU \
H0: μ_pobres >= μ_ricos \
Ha: μ_pobres < μ_ricos

```python
dfq2 = df.sort_values(by='Economy (GDP per Capita)')
df_pobres = dfq2.head(20)
df_ricos = dfq2.tail(20)
```

```python
#H0: μ_ricos <= μ_pobres 
#Ha: μ_ricos > μ_ricos
stats.ttest_ind(df_ricos['Happiness Score'], df_pobres['Happiness Score'], equal_var=True, alternative='greater')
```

```python
#H0: μ_pobres >= μ_ricos 
#Ha: μ_pobres < μ_ricos
stats.ttest_ind(df_pobres['Happiness Score'], df_ricos['Happiness Score'], equal_var=True, alternative='less')
```

Conclusão: \
Temos que pValue < 0.05. Portanto, rejeitamos a hipótese nula: os países com maior renda per capita tem maior índice de felicidade. Assim, dinheiro traz sim felicidade de acordo com este teste de hipótese.

Entre as provas foi oferecido um tutorial online de estatística. Realize um teste de hipótese que verifique se os alunos que realizaram o tutorial (online.tutorial = 1) tiveram uma melhora no desempenho na segunda avaliação (score2) em relação à primeira avaliação (score1).
```python
df = pd.read_csv('students.csv')
df.head()
```

### Questão 05
Deseja-se investigar se a religião do estudante tem algum impacto no seu desempenho acadêmico. Para tanto, realize um teste de hipótese nas notas do curso de estatística disponível no dataset 'students.csv' para as diferentes religões presentes no dataset. Para obter a nota final do curso, faça a média aritmética de score1 e score2 e salve em uma nova coluna chamada 'final.score'. Realize o teste com confiança de 95%.

H0: μ2 = μ3 = μ4 = μ4 = μ5 (média das notas são iguais) \
HA: médias das notas não são iguais 

```python
df['final.score'] = (df['score1'] + df['score2'])/2
df.head()
```

```python
df['religion'].unique()
```

```python
df1 = df[df['religion'] == 'Other']
df2 = df[df['religion'] == 'Orthodox']
df3 = df[df['religion'] == 'Catholic']
df4 = df[df['religion'] == 'Protestant']
df5 = df[df['religion'] == 'Muslim']
```

```python
stats.f_oneway(df1['final.score'], df2['final.score'],df3['final.score'],df4['final.score'],df5['final.score'])
```

Conclusão: \
Temos pValue > 0.05, portanto falhamos em rejeitar a hipótese nula. A média das notas não tem relação com a religião dos estudantes.

### Questão 06
Uma galeria de arte recebeu dois conjuntos de quadros (conjunto 1 e conjunto 2). O tamanho do quadro é definido pela medida da sua diagonal. Os arquivos 'quadros1.csv' e 'quadros2.csv' contém as medidas das diagonais dos conjuntos 1 e 2, respectivamente. De acordo com os dados contidos nesses arquivos, é possível dizer que o tamanho médio dos quadros dos dois conjuntos é o mesmo? Responda esta pergunta realizando um teste de hipótese.  Nível de confiança: 95%.

H0: μ1 = μ2 \
Ha: μ1 ≠ μ2

```python
df1 = pd.read_csv('quadros1.csv')
df2 = pd.read_csv('quadros2.csv')
stats.ttest_ind(df1, df2, equal_var=False)
```

Como o valor de pvalue = 0.03 é menor que 0.05, temos que rejeitar a hipótese nula. Assim, a média dos tamanhos dos quadros NÃO é a mesma.
